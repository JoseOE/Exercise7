```python
import matplotlib.pyplot as plt
```

```python
# Barras Valeria
marcas_clases=[1, 2, 3, 4, 5, 6]
frecuencias=[8, 5, 3, 8, 4, 2]
marcas_texto=["Pumas", "Chivas", "Santos", "America", "Toluca", "Necaxa"]
plt.figure(figsize = (12, 6))
plt.barh(marcas_clases, frecuencias, 
        height = 0.85, edgecolor = "k",
       color = ["#eb445b", "#f5938b", "#f0cdab", "#f1e7c5", "#CD617D", "#b6d4bb"])
plt.yticks(marcas_clases, marcas_texto, fontsize = 15)
plt.xlabel("Marcas de clase", fontsize = 20)
plt.ylabel("Frecuencias", fontsize = 20)
plt.title("Diagrama de Barras", fontsize = 25)
#plt.grid()
plt.show()
```

```python
datos = [26.67 , 16.67, 10, 26.67, 13.33, 6.66]
marcas_texto = ["Pumas", "Chivas", "Santos", "América", "Toluca", "Necaxa"]
separaciones = [0.1, 0, 0, 0.1, 0, 0]

plt.pie(datos, 
       explode = separaciones, 
       counterclock = False, 
       startangle = 90, 
       autopct = "%0.1f%%", 
       pctdistance = 0.8, 
       colors = ['#e4e6c9', '#e6dac6', '#d6c3b9', '#c2b48a', '#b37883', '#cdc7b5'], 
       labels = marcas_texto)
plt.show()
```
